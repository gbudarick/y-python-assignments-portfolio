```python
#part a

import sympy as sp
import numpy as np #originally used this for np.exp, but ended up using the sp.exp version
#function e^xsin(y)+y^3

x,y = sp.symbols ('x, y') #define x and y within sympy
f= sp.exp(x)*sp.sin(y)+(y**3) #write out function
df_dx= sp.diff(f, x) #prtial deriv with respect to x
df_dy= sp.diff(f, y) #partial deriv with respect to y

print ("partial derivative wrt x:", df_dx,"    ", "partial derivative wrt y:",df_dy)

```

```python
#part b

#starting variables
x0= 1
y0= -1
alpha= 0.01
num_iterations= 100

def gradient_descent(x0, y0,f, grad_f, alpha, num_iterations):
    x = x0
    y = y0 # Initialize x and y with the initial point
    
    for i in range(num_iterations):
    # obtain the gradient of f at (x, y)
        grad_x, grad_y =  grad_f(x,y)
    # Update x and y by taking a step in the opposite direction of the gradient
        x = x- (alpha*grad_x)
        y = y- (alpha*grad_y)
    return x, y

def f(x,y):
    f= ((x**2)*y)+(x*(y**2))


def grad_f(x,y):
    grad_x = 2*x*y + y**2
    grad_y = x**2 + 2*x*y

    return grad_x, grad_y

values= gradient_descent(x0, y0,f, grad_f, alpha, num_iterations)
print("Values of x and y:", values)

```

```python
#part c
#ln(x^2+y^2)

x,y = sp.symbols ('x, y') #define x and y within sympy
f= sp.ln((x**2)+(y**2)) #write out function
df_dx= sp.diff(f, x) #prtial deriv with respect to x
df_dy= sp.diff(f, y) #partial deriv with respect to y


df_dxx= sp.diff(df_dx, x) #second partial of x wrt x
df_dyy= sp.diff(df_dy, y) #second partial of y wrt y
df_dxy= sp.diff(df_dx, y) #partial of x wrt y
df_dyx= sp.diff(df_dy, x) #partial of y wrt x

print ("second partial derivative wrt x:", df_dxx,"    ", "second partial derivative wrt y:",df_dyy, "     ", "partial of x with respect to y:", df_dxy, "     ", "partial for y with respect to x:", df_dyx)
```

```python
#part d
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

x, y = sp.symbols('x y')
j = x**3 - 3*x*y + y**3

j_func = sp.lambdify((x, y), j, 'numpy')

x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = j_func(X, Y)

plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of $j(x, y) = x^3 - 3xy + y^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()

```

```python
#part e

import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

x, y = sp.symbols('x y') #choose a function, im going with 4y^5x+2x^2y^3/xy
j = y*sp.sin(x)+x*sp.cos(y)

j_func = sp.lambdify((x, y), j, 'numpy')

x_vals = np.linspace(-20, 20, 400)
y_vals = np.linspace(-20, 20, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = j_func(X, Y)

plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of $j(x, y) = ysin(x) +xsin(y) $')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```
