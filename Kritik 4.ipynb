```python
#Gradient descent is an algorithm to find critical points (local or absolute) 
#Based on the principle that the tangent at a critical point is 0 
#We want steps X0, ...., Xn such that f(Xn) is approx a local min
#so f'(Xn) = 0
#If f'(X0) > 0, then the funtion is increasing, as were looking to find a local min
#move to the left to get more negative and find minimum
#But by how much? Move some value eta >0, which we call the learning rate (how fast we
#converge on extrema), we assume eta is small, h= -eta f'(Xo), becoming f(Xo- eta f'(Xo))
#which is f'(Xo) (-eta f'(Xo)) + f(Xo), which is approz f(Xo) -eta (f'(Xo))^2

#From f(Xo- eta f'(Xo)) <= approx f(Xo), since eta is small and positive, so we now have
#an upper bount on z

#Upper bound is f(Xo-eta f'(X0)) 
```
